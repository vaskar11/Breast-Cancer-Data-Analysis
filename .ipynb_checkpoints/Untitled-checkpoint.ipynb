## outlier Detection

```python
import matplotlib.pyplot as plt
import numpy as np

# Select the columns for box plotting
names = df.columns[5:11]

# Collect the column values into a list
values = []
for column in df.iloc[:, 5:11].columns:
    li = df[column].tolist()
    values.append(li)

# Define the colors for each boxplot
colors = ['gold', 'mediumturquoise', 'darkorange', 'lightgreen', 'cyan', 'royalblue']

# Create a figure and axis for the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Create boxplot
box = ax.boxplot(values, patch_artist=True, notch=False, vert=True, labels=names)

# Customize each box plot with colors
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Customize the appearance of outliers and whiskers
for whisker in box['whiskers']:
    whisker.set(color='black', linewidth=1.5)

for cap in box['caps']:
    cap.set(color='black', linewidth=1.5)

for median in box['medians']:
    median.set(color='black', linewidth=2)

for flier in box['fliers']:
    flier.set(marker='o', color='red', alpha=0.5, markersize=5)

# Set axis labels
ax.set_title('Box Plot with Outliers for Selected Columns')
ax.set_ylabel('Values')

# Display the plot
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.tight_layout()
plt.show()
```

### data split

```python
random_state = 42
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df['diagnosis'], test_size = 0.2, random_state = random_state)
```

### feature engineering

#### scaler

In meachine learning, many algorithms (expect some algorithms ,like decision trees and random forests) are sensitive to features with different scales, so it’s important to scale numerical input variables to a standard range. Normalizer (standard deviation scaling) and StandardScaler (mean removal and standard deviation scaling) are the two most popular scaling techniques. In this case, we dectect in boxplots that some features contains a few extreme outliers which are not easy to remove or replace. Extreme outliers often influence the sample mean / variance in a negative way. Therefore, the above two scalers might not work very well here. We use RobustScaler as a replacement. This scaler is more robust to outliers by removing the median and scaling the data according to the quantile range.

```python
scale = RobustScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

### dimensionality reduction
From correlation matrix and Joint and marginal distributions plots, we observe that some features are highly correlated, like radius_mean, perimeter_mean and area_mean. Those features have almost the same effect on the dependent variable. In real world problem, classification algorithms including too many features in dataset commonly suffer from the "Curse of Dimensionality", which means more data, bulky computation, and also overfitting risk. Both feature selection (selecting specific features to include) and feature extraction(extracting a new feature set from the input features) can effectively avoid these problems. In practice, the effectiveness of dimensionality reduction highly depends on the algorithm to be applied later and type of the data to be passed into the methodology. Here, we will try two different methods: Principal component analysis (PCA) and Recursive features elimination(RFE), and also compare the results by separately using them on classification algorithms

#### principal component analysis(PCA)

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA

# Assuming X_train is already defined
pca = PCA()
pca.fit(X_train)
exp_var_cumul = np.cumsum(pca.explained_variance_ratio_)

# Create a figure
plt.figure(figsize=(10, 6))

# Plot the cumulative explained variance
plt.plot(np.arange(1, exp_var_cumul.shape[0] + 1), exp_var_cumul, marker='o', linestyle='-', color='blue')

# Add a horizontal dashed line at y=0.95
plt.axhline(y=0.95, color='red', linestyle='--', label='95% Explained Variance')

# Set labels and title
plt.title('Cumulative Explained Variance by Principal Components', fontsize=16)
plt.xlabel('# of Components', fontsize=14)
plt.ylabel('Cumulative Explained Variance', fontsize=14)
plt.xticks(np.arange(1, exp_var_cumul.shape[0] + 1))  # Ensure x-ticks are set correctly
plt.legend()  # Show the legend
plt.grid(True)  # Add grid lines for better readability

# Show the plot
plt.tight_layout()
plt.show()
```

 here we can see there is an elbow after the seventh component, and 91% of the total variance is explained by the firrst seventh components. If keeping the first 10 or 17 principal components, we can preserve about 95 % or even more than 99% of the total variance.

#### Recursive features elimination(RFE)
Recursive feature elimination (RFE) is a feature selection method. The process is that a separate model is trained and each time the weakest features are taken away, until the optimum set of features to use for the maximum results is reached.

```python
# Fit the RFE model to identify the optimum number of features .
rfecv = RFECV(cv=StratifiedKFold(n_splits=10, random_state=random_state, shuffle=True),
      estimator=DecisionTreeClassifier(), scoring='accuracy')
    
rfecv.fit(X_train, y_train)

print("Optimal number of features : %d" % rfecv.n_features_)

# Plot number of features VS. cross-validation scores
plt.figure(figsize=(15,5))
plt.xlabel("# of features selected")
plt.ylabel("Cross validation score (accuracy)")
plt.plot(
    range(1, len(rfecv.cv_results_['mean_test_score']) + 1),
    rfecv.cv_results_['mean_test_score'],
)
plt.show()
```

We can visualise how the model’s performance changes when the number of features increases by ploting the number of features used in the model versus the model accuracy. As we can see, more features isn’t always better. The accuracy value has reached the highest when the first # important features are used in the model.

```python
import pandas as pd

# Creating a list to store feature information
feature_list = []

# Identifying the features RFE selected
for i in range(X_train.shape[1]):
    row = {
        'feature': i,
        'support': rfecv.support_[i],
        'ranking': rfecv.ranking_[i]
    }
    feature_list.append(row)

# Create a DataFrame from the list
df_features = pd.DataFrame(feature_list)

# Sort the DataFrame by 'ranking' and display the top 10 features
df_features_sorted = df_features.sort_values(by='ranking')
print(df_features_sorted.head(10))  # You can also filter for supported features
```

### Grid search Cross Validation
#### find the best hyperparameter

```python
def modelselection(classifier, parameters, scoring, X_train):
    clf = GridSearchCV(estimator=classifier,
                   param_grid=parameters,
                   scoring= scoring,
                   cv=5,
                   n_jobs=-1)# n_jobs refers to the number of CPU's that you want to use for excution, -1 means that use all available computing power.
    clf.fit(X_train, y_train)
    cv_results = clf.cv_results_
    best_parameters = clf.best_params_
    best_result = clf.best_score_
    print('The best parameters for classifier is', best_parameters)
    print('The best training score is %.3f:'% best_result)
#    print(sorted(cv_results.keys()))
    return cv_results, best_parameters, best_result
```

```python
import numpy as np
import matplotlib.pyplot as plt

def PCA_curves(PCA_cv_score, PCA_test_score, training_time, n):
    fig, axs = plt.subplots(1, 2, figsize=(15, 5))

    # Plotting PCA CV Score and Test Score
    axs[0].plot(n, PCA_cv_score, color='purple', linewidth=2, label='CV score')
    axs[0].plot(n, PCA_test_score, color='cyan', linewidth=2, label='Test score')
    axs[0].set_title('# of Components in PCA versus Model Accuracy')
    axs[0].set_xlabel('# of components')
    axs[0].set_ylabel('Accuracy')
    axs[0].legend()
    axs[0].grid(True)

    # Plotting Training Time
    axs[1].plot(n, training_time, color='green', linewidth=2, label='Training time')
    axs[1].set_title('# of Components in PCA versus Training Time')
    axs[1].set_xlabel('# of components')
    axs[1].set_ylabel('Training time (seconds)')
    axs[1].legend()
    axs[1].grid(True)

    # Adjust layout
    plt.tight_layout()
    plt.show()

# Dummy Data for Testing
n = np.arange(1, 11)  # Number of components from 1 to 10
PCA_cv_score = np.random.rand(10)  # Random CV scores
PCA_test_score = np.random.rand(10)  # Random Test scores
training_time = np.random.rand(10) * 5  # Random training times in seconds

# Call the function with dummy data
PCA_curves(PCA_cv_score, PCA_test_score, training_time, n)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

def metrics(X, y_test, CV_clf):
    y_pred = CV_clf.predict(X)
    
    # Compute confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    tn = cm[0, 0]
    fp = cm[0, 1]
    fn = cm[1, 0]
    tp = cm[1, 1]
    
    Accuracy = (tp + tn) / (tp + tn + fp + fn)
    Sensitivity = tp / (tp + fn)
    Specificity = tn / (tn + fp)
    Precision = tp / (tp + fp)
    F_measure = 2 * tp / (2 * tp + fp + fn)
    
    print('Accuracy = %.3f' % Accuracy)
    print('Sensitivity = %.3f' % Sensitivity)  # same as recall
    print('Specificity = %.3f' % Specificity)
    print('Precision = %.3f' % Precision)
    print('F-measure = %.3f' % F_measure)

    # Plotting the confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                 xticklabels=['Predicted Negative', 'Predicted Positive'],
                 yticklabels=['Actual Negative', 'Actual Positive'])
    
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix')
    plt.show()
    
    return Accuracy, Sensitivity, Specificity, Precision, F_measure
```
